```python
import pandas as pd
df=pd.read_csv('train.csv')
X_train = df[['SL', 'EEG', 'HR']].values
Y_train=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
(trainData, testData, trainLabels, testLabels) = train_test_split(X_train,Y_train,test_size=0.25, random_state=42)
```

**Normalizamos los datos de entrenamiento y los de test**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
trainData = scaler.fit_transform(trainData)
testData = scaler.transform(testData)
```

# **Evaluación KNN**

**Primero hacemos el modelo y luego lo evaluamos**

```python
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
modelKNN = KNeighborsClassifier(n_neighbors=5)
modelKNN.fit(trainData, trainLabels)
```

```python
print("EVALUACIÓN EN CONJUNTO DE TEST USANDO KNN")
predictionsKNN = modelKNN.predict(testData)
print(classification_report(testLabels, predictionsKNN))
```

# **Evaluación Arboles**

**Primero hacemos el modelo y luego lo evaluamos**

```python
from sklearn.tree import DecisionTreeClassifier
modelArboles = DecisionTreeClassifier(min_samples_split=10)
modelArboles.fit(trainData, trainLabels)
```

```python
print("EVALUACIÓN EN CONJUNTO DE TEST USANDO ARBOLES")
predictionsArboles = modelArboles.predict(testData)
print(classification_report(testLabels, predictionsArboles))
```

# **Evaluación Regresion Logistica**

**Primero hacemos el modelo y luego lo evaluamos**

```python
from sklearn.linear_model import LogisticRegression
modelRegLog = LogisticRegression(solver="lbfgs",penalty="l2",C=1)
modelRegLog.fit(trainData, trainLabels)
```

```python
print("EVALUACIÓN EN CONJUNTO DE TEST USANDO REGRESION LOGISTICA")
predictionsRegLog = modelRegLog.predict(testData)
print(classification_report(testLabels, predictionsRegLog))
```

# **Evaluación SVM**

**Primero hacemos el modelo y luego lo evaluamos**

```python
from sklearn.svm import SVC
modelSVM = SVC(kernel='poly',gamma="scale",coef0=1)
modelSVM.fit(trainData, trainLabels)
```

```python
print("EVALUACIÓN EN CONJUNTO DE TEST USANDO SVM")
predictionsSVM = modelSVM.predict(testData)
print(classification_report(testLabels, predictionsSVM))
```

# **Evaluación Redes Neuronales**

**Primero hacemos el modelo y luego lo evaluamos**

```python
from sklearn.neural_network import MLPClassifier
modelRN = MLPClassifier(momentum=0.95,learning_rate_init=0.01,hidden_layer_sizes=(5,4,3,2),activation="tanh")
modelRN.fit(trainData, trainLabels)
```

```python
print("EVALUACIÓN EN CONJUNTO DE TEST USANDO REDES NEURONALES")
predictionsRN = modelRN.predict(testData)
print(classification_report(testLabels, predictionsRN))
```

# **RandomForest**

**Primero hacemos el modelo y luego lo evaluamos**

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(min_samples_split=10)
modelRF.fit(trainData, trainLabels)
```

```python
print("EVALUACIÓN EN CONJUNTO DE TEST USANDO RANDOM FOREST")
predictionsRF = modelRF.predict(testData)
print(classification_report(testLabels, predictionsRF))
```

**El mejor modelo es el de randomForest**
